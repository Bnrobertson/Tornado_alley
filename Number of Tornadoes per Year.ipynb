```python
import pandas as pd
from pathlib import Path
import seaborn as sns
```

```python
csv_path = Path("Resources/us_tornado_dataset_1950_2021.csv")
tornadoes_df = pd.read_csv(csv_path, encoding="UTF-8")
tornadoes_df.head(20)
```

```python
tornadoes_per_year = tornadoes_df["yr"].value_counts()
TPY_DF1 = pd.DataFrame(tornadoes_per_year)
Sorted_TPY_df = TPY_DF1.sort_values("yr")
Sorted_TPY_df.head()
```

```python
sns.displot(data=Sorted_TPY_df, x="yr", col="count", kde=True)
```
